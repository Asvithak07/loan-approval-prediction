```python
df=pd.read_csv("/content/loan_train.csv")
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df=df.drop("Loan_ID",axis=1)
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])

# Fill numerical missing values with median
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())

# Check if any missing values remain
print(df.isnull().sum())
```

```python
#encoding categorical columns

from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

cat_cols = ['Gender', 'Married', 'Dependents', 'Education',
            'Self_Employed', 'Property_Area', 'Loan_Status']

le = LabelEncoder()
for col in cat_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

print(df_encoded.head())
```

```python
df_encoded.to_csv("cleaned_dataset_test|.csv", index=False)

print("✅ Cleaned dataset saved as 'cleaned_dataset.csv'")
```

```python
train_df = pd.read_csv("/content/cleaned_dataset.csv")

# Separate features (X) and target (y)
X_train = train_df.drop(columns=["Loan_Status"])  # replace "target" with your column name
y_train = train_df["Loan_Status"]
```

```python
from sklearn.model_selection import train_test_split

X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train_split, y_train_split)
```

```python
from sklearn.metrics import accuracy_score

y_pred_val = model.predict(X_val_split)
print("Validation Accuracy:", accuracy_score(y_val_split, y_pred_val))
```

```python
test_df = pd.read_csv("/content/cleaned_dataset_test|.csv")

if "Loan_Status" in test_df.columns:
    test_df = test_df.drop("Loan_Status", axis=1)

# Make predictions
test_predictions = model.predict(test_df)

# Save predictions
pd.DataFrame({"Prediction": test_predictions}).to_csv("test_predictions.csv", index=False)
```

```python
import pandas as pd

pred_df = pd.DataFrame({"Prediction": test_predictions})
pred_df["Prediction_Label"] = pred_df["Prediction"].map({0: "Rejected", 1: "Approved"})

pred_df.to_csv("test_predictions.csv", index=False)
```

```python
#evaluating the model

from sklearn.metrics import classification_report, confusion_matrix

y_true = pd.read_csv("/content/cleaned_dataset_test|.csv")["Loan_Status"]
y_pred = model.predict(test_df)

print(classification_report(y_true, y_pred))
print(confusion_matrix(y_true, y_pred))
```

```python
#saving the trained model so that you don't need to retrain the model

import joblib

joblib.dump(model, "loan_approval_model.pkl")
```
